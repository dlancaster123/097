```python
import pandas as pd
import plotly.express as px  # Import Plotly Express
import matplotlib.pyplot as plt

url = 'https://docs.google.com/spreadsheets/d/1FR7m8CjOp7Ch9fHPDTQGipIAS9XZPqFmK9iXg7b0EU0/export?format=csv'
df = pd.read_csv(url, header=2)

def piecharts(subset_df, input1, input2):
    titles = ['Base', 'Signature', 'Elite']
    for i, title in enumerate(titles):
        fig = px.pie(subset_df, values=title, names=subset_df['Category'], title=f'{title} Level Shelves in {input1} Region')

        fig.show()

def nonswire(subset_df, input1):
    print(f"\nSo what you will see is the first chart is just showing how the companies should break down their shelves, regardless of Coca-Cola products. Then the three following charts are showing what percentage of the broken down percentage (first chart) Swire is asking for.")
    fig = px.pie(subset_df, values='Shelves', names=subset_df['Category'], title=f'Shelves in {input1} Region (Not Swire)')
    fig.show()

def rows(input1, input2):
    input1 = input1.lower()
    input2 = input2.lower()
    if input1 == "az and nm" and input2 == "no":
        return df[0:10]
    elif input1 == "oregon" and input2 == "no":
        return df[10:20]
    elif input1 == "oregon" and input2 == "yes":
        return df[20:30]
    elif input1 == "rm" and input2 == "yes":
        return df[30:40]
    elif input1 == "rm" and input2 == "no":
        return df[40:50]
    elif input1 == "northern co" and input2 == "no":
        return df[50:60]
    elif input1 == "northern co" and input2 == "yes":
        return df[60:70]
    elif input1 == "southern co" and input2 == "yes":
        return df[70:80]
    elif input1 == "utah" and input2 == "no":
        return df[80:90]
    elif input1 == "utah" and input2 == "yes":
        return df[90:100]
    elif input1 == "washington" and input2 == "no":
        return df[100:110]
    elif input1 == "eastern wa" and input2 == "no":
        return df[110:120]
    elif input1 == "reno" and input2 == "no":
        return df[120:130]
    else:
        return pd.DataFrame()

def allpie():
    regions = [("az and nm", "no"), ("oregon", "no"), ("oregon", "yes"), ("rm", "yes"), ("rm", "no"), ("northern co", "no"), ("northern co", "yes"), ("southern co", "yes"), ("utah", "yes"), ("utah", "no"), ("washington", "no"), ("eastern wa", "no"), ("reno", "no")]
    for region, condition in regions:
        subset_df = rows(region, condition)
        if not subset_df.empty:
            nonswire(subset_df, region)
            piecharts(subset_df, region, condition)

while True:
    input1 = input(f"What region would you like to see? Say 'quit' if you don't wish to continue: ")
    if input1.lower() == "quit":
        break
    input2 = input(f"Is it KO and DP? ")
    if input1.lower() == "all":
        allpie()
    else:
        subset_df = rows(input1, input2)
        if subset_df.empty:
            print("Not Valid")
            continue
        nonswire(subset_df, input1)
        piecharts(subset_df, input1, input2)
        question = input(f"Would you like to see another region? ")
        if question.lower() == "no":
            print("I hope this has helped you visualize the information in the table!")
            break

print(f"As you can see, some regions we are slacking and we aren't asking for much shelf space. As we assess our numbers and create a plan for 2025, we should see what we can do to increase our product sales in these areas.")

```
